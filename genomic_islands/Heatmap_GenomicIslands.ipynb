```python
%matplotlib inline
from collections import defaultdict
```

```python
#Create a table (and maybe a plot) for the genes present on the genomes island in Shigella
```

```python
# First, let's import the table and store in a dictionary

genome_info = list() # Genome information list
freq_info = list() # Frequency information dictionary

genome_list = ["a0173","a0231","a0144","b0200","c8405","c8505"]

combined_data_dict = defaultdict() # Combined data used to generate the new table
combined_data_list = list() # To generate plots

# Just to keep the gene order (dictionary change the order of keys), I will
# make a list of the gene ids to use
gene_id = list()

# Store data from genome info
for line in open("orgs_with_genes.txt", 'r'):
    if not line.strip(): # Skip empty lines
        continue
    line = line.rstrip() # Remove ends of line from the file
    genome_info.append(line.split("\t")[1:])

# Store data from freqs
for line in open("n_genes_orgs.txt", 'r'):
    if not line.strip():
        continue
    line = line.rstrip()
    freq_info.append(line.split("\t")[1:])
    gene_id.append(line.split("\t")[0])

# Very ugly here :(
for gene_in_genome, gene_frequency, id in zip(genome_info, freq_info,gene_id):
    count = defaultdict()
    for entry in genome_list:
        count[entry] = 0
    
    for genome, gene_count in zip(gene_in_genome, gene_frequency):
        count[genome] = gene_count
    
    combined_data_dict[id] = count

```

```python
# Output a new table
output_file = open("output_table.txt", 'w')

output_file.write("GeneID_Uniprot\t")
output_file.write("\t".join(genome_list) + "\n")

for entry in gene_id:
    output_file.write(entry + "\t")
    
    output_values = list()
    
    for genome in genome_list:
        output_values.append(combined_data_dict[entry][genome])
    
    output_file.write("\t".join(map(str,output_values)) + "\n")

output_file.close()
```

```python
#Add annotation
import urllib
import time

annotation = defaultdict()
for uniprot_id in gene_id:
    time.sleep(1)
    data = urllib.urlopen("http://www.uniprot.org/uniprot/" + uniprot_id + ".txt")
    
    for line in data:
        if line.startswith("DE"):
            annotation[uniprot_id] = line
```

```python
annotation_file = open("output_annotation.txt", 'w')

for gene in gene_id:
    annotation_file.write(gene + "\t")
    annotation_file.write(annotation[gene])

annotation_file.close()
```

```python
# Plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("output_table.txt", sep="\t", index_col=0)
```

```python
f, ax = plt.subplots(figsize=(20,25))
ax.tick_params(labelsize=13)
sns.heatmap(df, annot=True, square=True)
f.tight_layout()
f.savefig("gene_name_heatmap.pdf")
```

```python
f, ax = plt.subplots(figsize=(20,25))
ax.tick_params(labelsize=13)
sns.heatmap(df, annot=True, square=True, yticklabels=annotation.values())
f.tight_layout()
f.savefig("function_heatmap.pdf")
```
